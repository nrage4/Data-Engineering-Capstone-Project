# 2)	Provide information on user demographics (age range, region) would help them hone in on marketing spend



```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df['uk_region'] = df['uk_region'].astype('category') 
```

```python
# this shows the pie chart of the different UK regions and the ratio between them 

# separating the highest % from the others by separating
explode = (0 ,0,0.2)

# using seaborn for changing colour
colors = sns.color_palette('viridis')

# changing the size of the pie chart
fig = plt.figure(figsize =(11,9)) 

# adding the label
labels= [' UK North', ' UK North east', 'UK South east']

# plotting the pie chart

df.groupby(['uk_region'])['quantity'].sum().plot(kind='pie', ylabel='',
                                                 autopct='%1.0f%%', explode=explode,
                                                 colors=colors, shadow=True, labels=None)

plt.title('Region demographics', weight = 'bold', fontsize = 14)
plt.legend(bbox_to_anchor=(0.8 ,1), labels=labels)
plt.savefig('Region_Demographics.png', bbox_inches='tight', dpi= 150, facecolor=(1,1,1))
plt.show()



          
```


This pie chart displays the geographic origin of the majority of the customers who purchased toys and toothbrushes. The pie chart reveals that UK South East is where the majority of the customers reside. 

```python

df.pivot_table(index='uk_region', columns = 'product', aggfunc=np.sum ).plot(kind='bar', 
                                                         y='quantity', edgecolor='grey', width = 0.8, 
                                                         ylabel= 'Quantity'       
                                                          )

plt.title('Quantity of products bought by UK region', weight = 'bold', fontsize = 14)
 
plt.legend(loc='upper right', bbox_to_anchor=(1.35, 1))
plt.savefig('Total_Quantity_Region.png', bbox_inches='tight', dpi= 150, facecolor=(1,1,1))

plt.show()
```

It is clear from the bar chart above how many toothbrushes and toys were purchased in each region. The only region to purchase toys was the UK South East. Whereas toothbrushes were purchased in all three regions.

```python
#creating line graph to provide information on the age range of who bought the products
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize =(10,7)) # changing the size
plt.plot(df.groupby(['age'])['quantity'].sum(), color='red', marker='o') 
plt.title('Age Demographics', fontsize=14, weight = 'bold') 
plt.xlabel('Age')
plt.ylabel('Quantity')
plt.savefig('Age_Demographics.png', bbox_inches='tight', dpi= 150, facecolor=(1,1,1))
plt.grid(True)
plt.show()
```

 
The line graph above demonstrates how different age groups purchased toothbrushes or toys. For instance, 12 products have been purchased by consumers who are 35 years old, but just one product has been purchased by a customer who is 45 years old.

```python
df.pivot_table(index='product', columns='month_order', aggfunc=np.sum ).plot(kind='bar', 
                                                         y='quantity', edgecolor='grey', width = 0.8, 
                                                         ylabel= 'Quantity' ,
                                                         align = 'center' 
                                                          )

plt.title('Total products bought for toothbrushes and toys', weight = 'bold', fontsize = 14)

plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.savefig('Products_Sold_By_Month.png', bbox_inches='tight', dpi= 150, facecolor=(1,1,1))
plt.show()


```

The bar graph above shows that sales of toothbrushes are generally stable, with the exception of may, when they are relatively low and there were no sales in January. In terms of sales of toys, January had the highest volume and May and December saw none.

Since there are less sales in the summer than there are during the winter, I can see from this data that it is not consistent throughout the year and they are seasonal as from Dec-March have higher sales compared to April-September where there are barely any sales.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

# importing the html index

```python
URL = "https://mystaticwebsite-3.s3.amazonaws.com/index.html"
page = requests.get(URL)
print(page.text)
```

```python

soup = BeautifulSoup(page.content, "html.parser")

results1 = soup.find(id="south_east")
results2= soup.find(id="north_east")
results3 = soup.find(id="north")
results4 = soup.find(id="south")

```

```python
#cresting a table from the hmtl information
marketing_table = soup.table
df_marketing = pd.read_html(str(marketing_table))[0]
```

```python
df_marketing
```

```python
#finding the mean for cpa in by uk region - series
df.groupby(['uk_region'])['cpa'].mean() 
```

```python
#changing the column name to uk_region from Region to have the 
#same column name as our original df
df_marketing = df_marketing.rename(columns={"Region":"uk_region"})
df_marketing
```
